#### This notebooks splits Ghazi data into five fold for cross-validation training

```python
import pickle
import csv
```

```python
#load full data
pkl_file = open('eca_sti_data.pkl', 'rb')
data_all = pickle.load(pkl_file)
pkl_file.close()
```

```python
for data in data_all:
    if 'Feeling a little frightened ' in data['text']:
        print(data)
```

```python
# load dev split data
pkl_dev_file = open('eca_sti_dev_ids.pkl', 'rb')
dev_ids = pickle.load(pkl_dev_file)
pkl_dev_file.close()
```

```python
def getIndex(lst1, lst2):
    return next(((i, i + (len(lst2) if len(lst2) > 1 else 0)) for i in range(len(lst1) - len(lst2) + 1) if lst1[i:i + len(lst2)] == lst2), None)
```

```python
import string
folds = ['fold0', 'fold1', 'fold2', 'fold3', 'fold4']
for split_idx, dev_split in enumerate(dev_ids):
    train_split = set(list(range(len(data_all)))) - set(dev_split)
    train_split = list(train_split)
    with open('./{}/dev.tsv'.format(folds[split_idx]), 'wt') as out_file:
        tsv_writer = csv.writer(out_file, delimiter='\t')
        tsv_writer.writerow(['document', 'token_label', 'emotion-label'])
        for id in dev_split:
            data_point = data_all[id]
            document = data_point['text']
            print(data_point['id'])
            print(document)
            cause = data_point['cause']
            # if cause[-1] in string.punctuation:
            #     cause = cause[:-1]
            print(cause)
            print('*' * 50)
            emotion = data_point['emotion']
            token_label = []
            start_idx, end_idx = getIndex(document.split(), cause.split())  
            for idx, word in enumerate(document.split()):
                if idx == start_idx:
                    token_label.append('B-CAU')
                elif idx > start_idx and idx <= end_idx - 1:
                    token_label.append('I-CAU')
                else:
                    token_label.append('O')
            
            assert len(document.split()) == len(token_label)
            tsv_writer.writerow([document, ' '.join(token_label), emotion])

    with open('./{}/train.tsv'.format(folds[split_idx]), 'wt') as out_file:
        tsv_writer = csv.writer(out_file, delimiter='\t')
        tsv_writer.writerow(['document', 'token_label', 'emotion-label'])
        for id in train_split:
            data_point = data_all[id]
            document = data_point['text']
            cause = data_point['cause']
            # if cause[-1] in string.punctuation:
            #     cause = cause[:-1]
            emotion = data_point['emotion']
            token_label = []
            start_idx, end_idx = getIndex(document.split(), cause.split())
            for idx, word in enumerate(document.split()):
                if idx == start_idx:
                    token_label.append('B-CAU')
                elif idx > start_idx and idx <= end_idx - 1:
                    token_label.append('I-CAU')
                else:
                    token_label.append('O')
            
            assert len(document.split()) == len(token_label)
            tsv_writer.writerow([document, ' '.join(token_label), emotion])
```
